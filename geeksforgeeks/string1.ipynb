```python
#1
#palindrome string

class Solution:
    def isPalindrome(self, s):
        # code here
        left=0
        right=len(s)-1
        while left<right:
            if s[left]!=s[right]:
                return False
            left+=1
            right-=1
            
        return True
```

```python
#2
#Anagrams

class Solution:
    def areAnagrams(self, s1, s2):
       # code here
        dic={}
        
        if len(s1)!=len(s2):
            return False
        
        for i in s1:
            dic[i]=dic.get(i,0)+1
            
        for i in s2:
            if dic.get(i,0)==0:
                return False
            dic[i]-=1
       
        return True
        
  
```

```python
#3
# Remove the vowels in a string

class Solution:
	def removeVowels(self, s):
		# code here
		v=['a','e','i','o','u']
		result=''
		
		for i in s:
		    if i not in v:
		        result+=i
		
		return result
		        
```
