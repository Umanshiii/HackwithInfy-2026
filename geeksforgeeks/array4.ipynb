```python
#1
#Max Sum Subarray of size K

class Solution:
    def maxSubarraySum(self, arr, k):
        # code here 
        count=0
        
        for i in range(k):
            count+=arr[i]
        
        maxsum=count
        
        for i in range(k,len(arr)):
            count+=arr[i]
            count-=arr[i-k]
            if count>maxsum:
                maxsum=count
            
        return maxsum
```

```python
#2
#Count distinct elements in every window

class Solution:
    def countDistinct(self, arr, k):
        # Code here
        
        count=[]
        temp={}
        
        for i in range(k):
            temp[arr[i]]=temp.get(arr[i],0)+1
            
        count.append(len(temp))
        
        for i in range(k, len(arr)):
            
            temp[arr[i]]=temp.get(arr[i],0)+1
            temp[arr[i-k]]-=1
            
            if temp[arr[i - k]] == 0:
                del temp[arr[i - k]]
                
            count.append(len(temp))
        
        return count
```

```python
#3
#First negative integer in every window of size k

class Solution:
    def firstNegInt(self, arr, k): 
         # code here 
         
        res=[]
        window=[]
         
        for i in range(k):
            if arr[i] < 0:
                window.append(i)
        
        if len(window)!=0:
            res.append(arr[window[0]])
        else:
            res.append(0)
        
        for i in range(k,len(arr)):
            if window and window[0]<=i-k:
                window.pop(0)
            if arr[i]<0:
                window.append(i)
            if len(window)!=0:
                res.append(arr[window[0]])
            else:
                res.append(0)
            
        return res
```

```python
#4
# Trapping Rain Water

class Solution:
    def maxWater(self, arr):
        # code here
        left=0
        right=len(arr)-1
        leftmax=arr[0]
        rightmax=arr[len(arr)-1]
        water=0
        
        while left<right:
            
            if arr[left]<arr[right]:
                if arr[left]<leftmax:
                    diff=leftmax-arr[left]
                    water+=diff
                else:
                    leftmax = arr[left] 
                left += 1
            
            else:
                if arr[right]<rightmax:
                    diff=rightmax-arr[right]
                    water+=diff
                else:
                    rightmax= arr[right]
                right-=1
                
        return water
        
        
        
```
