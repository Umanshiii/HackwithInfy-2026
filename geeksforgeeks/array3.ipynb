```python
#1
#Frequencies in a Limited Array

class Solution:
    def frequencyCount(self, arr):
        #  code here
        n=len(arr)
        result=[0]*n
        
        for i in arr:
            if 1 <= i <= n:
                result[i-1]+=1
            
        return result
```

```python
#2
# Non Repeating Element in an Array

class Solution:
    def firstNonRepeating(self, arr): 
        # code here
        res={}
        
        for i in range(len(arr)):
            res[arr[i]] = res.get(arr[i],0) + 1
        
        for i in range(len(arr)):
            if res[arr[i]]==1:
                return arr[i]
                
        return 0
```

```python
#3
# remove duplicate from unsorted array

class Solution:
    def removeDuplicate(self, arr):
        # code here
        seen=[False]*100001
        res=[]
        for i in arr:
            if seen[i]==False:
                res.append(i)
                seen[i]=True
                
        return res
    
```

```python
#4
#Replace elements by rank

class Solution:
    def replaceWithRank(self, arr):
        # code here
        lis=[]
        
        for i in range(len(arr)):
            lis.append((arr[i],i))
            
        lis.sort()
        for i,(val,index) in enumerate(lis):
            arr[index]=i
```

```python
#5
#kadanes Algorithm

class Solution:
    def maxSubarraySum(self, arr):
        # Code here
        maximum=arr[0]
        currmax=arr[0]
        for i in range(1,len(arr)):
            currmax=max(arr[i],currmax+arr[i])
            if currmax>maximum:
                maximum=currmax
            
        return maximum
```
