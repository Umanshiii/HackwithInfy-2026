```python
#1
#Remove string duplicates

class Solution:
	def removeDuplicates(self, s):
	    # code here
	    seen=set()
	    output=''
	    for i in s:
	        if i not in seen:
	            output+=i
	            seen.add(i)
	            
	    return output
```

```python
#2
# Most frequent character in a string

class Solution:
    def getMaxOccurringChar(self, s):
        #code here
        dic={}

        for i in s:
            dic[i]=dic.get(i,0)+1

        max_val= max(dic.values())

        return min(k for k, v in dic.items() if v == max_val)
```

```python
#3
#Roman to Integer
class Solution:
    def romanToDecimal(self, s): 
        # code here
        dic = {'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000}
        
        out = dic[s[-1]]
        
        i = len(s) - 2
        
        while i >= 0:
            if dic[s[i]] < dic[s[i+1]]:
                out -= dic[s[i]]
            else:
                out += dic[s[i]]
            
            i -= 1
            
        return out
```

```python
#4
#Integer to Roman

class Solution:
    def convertRoman(self, n):
        #Code here
        
        roman=[(1,'I'),(4,'IV'),(5,'V'),(9,'IX'),(10,'X'),(40,'XL'),(50,'L'),(90,'XC'),(100,'C'),(400,'CD'),(500,'D'),(900,'CM'),(1000,'M')]
        roman.sort(reverse=True)
        res=''
        
        for i in roman:
            while n>=i[0]:
                res+=i[1]
                n-=i[0]
                
        return res
```

```python
#5
# Check for Subsequence

class Solution:
    def isSubSeq(self, s1, s2):
        # code here
        if len(s2)<len(s1):
            return False
        i=0
        j=0
        while i<len(s1) and j<len(s2):
            if s1[i]==s2[j]:
                i+=1
            j+=1
        
        if i==len(s1):
            return True
        else:
            return False
            
```
